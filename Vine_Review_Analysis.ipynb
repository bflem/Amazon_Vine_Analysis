```python
import pandas as pd
```

```python
# Read in the data
vine_table = pd.read_csv("vine_table.csv")
vine_df = pd.DataFrame(vine_table)
vine_df

# Add column names
vine_df.columns =['review_id', 'star_rating', 'helpful_votes', 'total_votes', 'vine', 'verified_purchase']
vine_df
```

```python
# Filter the data for more than 20 total votes
filtered_vine_df = vine_df[vine_df["total_votes"] >= 20]
filtered_vine_df
```

```python
# Filter new dataframe for rows where helpful votes / total votes >= 50%
helpful_vine_df = filtered_vine_df[filtered_vine_df["helpful_votes"] / filtered_vine_df["total_votes"] >= 0.5]
helpful_vine_df
```

```python
# Filter new dataframe for "y" in vine column
vine_yes_df = helpful_vine_df[helpful_vine_df["vine"] == "Y"]
vine_yes_df
```

```python
# Filter new dataframe for "n" in vine column
vine_no_df = helpful_vine_df[helpful_vine_df["vine"] == "N"]
vine_no_df
```

```python
# Find total number of reviews
index = helpful_vine_df.index
number_of_rows = len(index)

print(number_of_rows)
```

```python
# Find number of 5 star reviews
five_star_reviews = helpful_vine_df[helpful_vine_df["star_rating"] == 5]
len(five_star_reviews)
```

```python
# Find percentage of 5 star reviews for paid reviews
num_five_reviews_paid = len(vine_yes_df[vine_yes_df["star_rating"] == 5])
paid_percentage = num_five_reviews_paid/number_of_rows
paid_percentage
```

```python
# Find percentage of 5 star reviews for unpaid reviews
num_five_reviews_unpaid = len(vine_no_df[vine_no_df["star_rating"]== 5])
unpaid_percentage = num_five_reviews_unpaid/number_of_rows
unpaid_percentage
```
